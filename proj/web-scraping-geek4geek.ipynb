```python
from bs4 import BeautifulSoup as bs
import requests


page = requests.get("https://en.wikipedia.org/wiki/Main_Page")

soup = bs(page.content,'html.parser')
print(soup.prettify())
```

```python
list(soup.children)
```

```python
#find all occurences of p in HTML
#includes HTML tags
print(soup.find_all('p'))

print('\n\n')
```

```python
#return only text, does not include HTML tags
print(soup.find_all('p')[0].get_text())
```

```python
#return only text, does not include HTML tags
print(soup.find_all('p')[0].get_text())
```

```python
for i in range(11):
    print(soup.find_all('p')[i].get_text())
```

```python
#create object
object = soup.find(id='mp-left')

#find tags
items = object.find_all(class_='mp-h2')
result = items[0]

print(result.prettify())
```
